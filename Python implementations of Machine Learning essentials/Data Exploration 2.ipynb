## Plotting the data using a scatter plot with weight as a function of height

```python
_ = df.plot(kind='scatter', x='Height', y='Weight')
```

## Plotting the male and female poplulations with two different colors

```python
male = df[df['Gender'] == 'Male']
female = df.query('Gender == "Female"')
fig, ax = plt.subplots()

male.plot(kind='scatter', x='Height', y='Weight', ax=ax, color = 'blue',
           alpha = 0.2, title='Male and Female Populations')

female.plot(kind = 'scatter', x = 'Height', y = 'Weight',
            ax = ax, color = 'red', alpha = 0.2)

ax.legend(['male', 'female'])
```

```python
df['Gender Color'] = df['Gender'].map({'Male' : 'blue', 'Female': 'red'})
df.head()
```

```python
df.plot(kind = 'scatter',
       x = 'Height',
       y = 'Weight',
       c = df['Gender Color'],
       alpha = 0.3,
       title = 'Male and Female Population')
```

```python
fig, ax = plt.subplots()
ax.plot(male['Height'], male['Weight'], 'ob',
       female['Height'], female['Weight'], 'or', alpha = 0.3)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Male and Female Populations')
```
