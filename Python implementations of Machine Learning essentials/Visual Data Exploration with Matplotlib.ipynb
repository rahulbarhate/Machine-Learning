```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = np.random.normal(0, 0.1, 1000)
data2 = np.random.normal(1, 0.4, 1000) + np.linspace(0,1,1000)
data3 = 2+np.random.random(1000) + np.linspace(1,5,1000)
data4 = np.random.normal(3, 0.2, 1000) + 0.3 * np.sin(np.linspace(0,20,1000))
```

#### Stacking the data onto a single array

```python
data = np.vstack([data1, data2, data3, data4]).transpose()
```

```python
df = pd.DataFrame(data, columns = ['data1', 'data2', 'data3', 'data4'])
```

```python
df.head()
```

```python
df.info()
```

## Line Plot

```python
df.plot(title = 'Line Plot')
```

```python
#In order to plt using matplotlib funtions parameters have t be passed manually

plt.plot(df)
plt.title('Line Plot')
plt.legend(['data1','data2','data3','data4'])
```

## Scatter Plot

```python
df.plot(style = '.')
```

```python
_ = df.plot(kind = 'scatter', x = 'data1', y = 'data2', xlim = (-1.5, 1.5), ylim = (0, 3))
```

## Histograms

```python
df.plot(kind='hist',
       bins=50,
       title='Histogram',
       alpha=0.6,
       figsize=(5,5))
```

## Cumulative Distibution

```python
df.plot(kind='hist',
       bins=100,
       title='Cumulative distributions',
       normed = True,
       cumulative = True,
       alpha = 0.4)
```

## Box Plot

```python
df.plot(kind = 'box', title = 'Boxplot')
```

## Subplots

```python
fig, ax = plt.subplots(2,2, figsize=(5,5))

df.plot(ax = ax[0][0],
       title = 'Line Plot')

df.plot(ax = ax[0][1],
       style = 'o',
       title = 'Scatter Plot')

df.plot(ax=ax[1][0],
       kind = 'hist',
       bins = 50,
       title = 'Histogram')

df.plot(ax = ax[1][1],
       kind = 'box',
       title = 'Boxplot')

# Styling
plt.tight_layout()
```

## Pie Charts

```python
# How many data values have the count greater than 0.1

greaterthan01 = df['data1'] > 0.1
piecounts = greaterthan01.value_counts()

piecounts
```

```python
piecounts.plot(kind='pie',
              figsize = (5,5),
              explode = [0, 0.15],
              labels = ['<= 0.1', '>0.1'],
              autopct = '%1.1f%%',
              shadow = True,
              startangle = 90,
              fontsize = 16)
```

## Hexbin Plot - Useful to show 2D Distributions

```python
data = np.vstack([np.random.normal((0,0), 2, size=(1000,2)),
                 np.random.normal((9,9), 3, size=(2000,2))])

df = pd.DataFrame(data, columns = ['x', 'y'])
```

```python
df.head()
```

```python
df.plot()
```

```python
df.plot(kind = 'hexbin', x='x', y='y', bins = 100, cmap = 'rainbow')
```
